# Variables Aleatorias

### Definicion

En probabilidad y estadística, una **variable aleatoria** (v.a.) es una función que asocia a cada resultado de un experimento aleatorio un número real.

Formalmente, si $(\Omega, \mathcal{F}, P)$ es un espacio de probabilidad, una variable aleatoria es una función:

$$
X : \Omega \longrightarrow \mathbb{R}
$$

de modo que para cualquier número real $x$, el conjunto $\{\omega \in \Omega : X(\omega) \le x\}$ pertenece a $\mathcal{F}$.

En términos prácticos, una variable aleatoria **traduce resultados experimentales** en valores numéricos que pueden analizarse mediante métodos matemáticos y estadísticos.


### Tipos de Variables Aleatorias

Existen dos tipos fundamentales:

1. **Variable aleatoria discreta:**  
   Toma un número **finito o numerable** de valores posibles.
   Ejemplo: número de caras al lanzar tres monedas.

2. **Variable aleatoria continua:**  
   Puede tomar **cualquier valor dentro de un intervalo** de números reales.
   Ejemplo: tiempo que tarda un autobús en llegar.


## Variables Aleatorias Discretas

#### Definicion

Una **variable aleatoria discreta** (v.a.d.) es una función que asigna un número real a cada resultado posible de un experimento aleatorio, tomando únicamente **valores finitos o numerables**.

Formalmente, si $(\Omega, \mathcal{F}, P)$ es un espacio de probabilidad, una variable aleatoria discreta es una función:

$$
X : \Omega \longrightarrow \mathbb{R}
$$

tal que el conjunto de valores que puede tomar $X$ es **finito o infinito numerable**:

$$
R_X = \{x_1, x_2, x_3, \dots \}
$$


#### Propiedades de la función de probabilidad

Toda función $f_X(x)$ debe cumplir:
1. **No negatividad:**
   $$
   f_X(x) \ge 0, \quad \forall x
   $$

2. **Normalización:**
   $$
   \sum_{x_i \in R_X} f_X(x_i) = 1
   $$

3. **Probabilidad de un evento:**
   $$
   P(a \le X \le b) = \sum_{x_i \in [a,b]} f_X(x_i)
   $$


#### Medidas asociadas

- **Esperanza o valor esperado:**
  $$
  \mathbb{E}[X] = \sum_{x_i} x_i \, f_X(x_i)
  $$

- **Varianza:**
  $$
  \text{Var}(X) = \sum_{x_i} (x_i - \mathbb{E}[X])^2 f_X(x_i)
  $$

Estas medidas describen el **comportamiento promedio** y la **dispersión** de la variable aleatoria.


#### Ejemplo — Variable aleatoria del lanzamiento de un dado

Sea $X$ la variable aleatoria que representa el número obtenido al lanzar un dado justo.

Los posibles valores de $X$ son:
$$
R_X = \{1, 2, 3, 4, 5, 6\}
$$

Como todos los resultados son equiprobables:
$$
f_X(x) =
\begin{cases}
\frac{1}{6}, & x = 1, 2, 3, 4, 5, 6 \\
0, & \text{en otro caso.}
\end{cases}
$$

**Comprobación de normalización:**
$$
\sum_{x=1}^{6} f_X(x) = 6 \cdot \frac{1}{6} = 1
$$

**Esperanza y Varianza:**
$$
\mathbb{E}[X] = \sum_{x=1}^{6} x f_X(x) = \frac{1+2+3+4+5+6}{6} = 3.5
$$

$$
\text{Var}(X) = \sum_{x=1}^{6} (x - 3.5)^2 \frac{1}{6} = \frac{35}{12} \approx 2.9167
$$

**Interpretación:**  
El resultado esperado al lanzar un dado justo es **3.5**, y la varianza mide la dispersión respecto a ese valor promedio.

La representación gráfica de una variable discreta se realiza con un **diagrama de barras**, donde cada barra indica la probabilidad de cada valor posible de $X$.

En este ejemplo, las seis barras tienen la misma altura, reflejando la equiprobabilidad del dado.


```python
## Variable Aleatoria Discreta — Ejemplo: Dado justo
## Visualización de la función de masa de probabilidad (f.m.p.),
## y cálculo de esperanza y varianza.

import numpy as np
import matplotlib.pyplot as plt

# Definición de los valores posibles de X
x = np.arange(1, 7)
p = np.full(6, 1/6)  # Probabilidades iguales

# Esperanza y varianza
esperanza = np.sum(x * p)
varianza = np.sum((x - esperanza)**2 * p)

# Resultados
print(f"Esperanza (E[X]) = {esperanza:.2f}")
print(f"Varianza (Var[X]) = {varianza:.4f}")

# Gráfica de la f.m.p.
plt.figure(figsize=(6, 4))
plt.bar(x, p, color='steelblue', edgecolor='black')
plt.title("Función de Masa de Probabilidad — Dado Justo", fontsize=12)
plt.xlabel("Valor de X")
plt.ylabel("f_X(x) = P(X = x)")
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

### Uniforme Discreta


#### Definición

Una **variable aleatoria uniforme discreta** es aquella que puede tomar un número **finito de valores distintos**, todos con la **misma probabilidad**.

Este tipo de variable describe experimentos donde cada resultado posible es **equiprobable**, es decir, no hay preferencia por ningún valor.

Sea una variable aleatoria $X$ que puede tomar $n$ valores distintos:

$$
X \in \{x_1, x_2, \dots, x_n\}
$$

Se dice que $X$ tiene una **distribución uniforme discreta** si:

$$
P(X = x_i) = \frac{1}{n}, \quad i = 1, 2, \dots, n
$$

o equivalentemente,

$$
f_X(x) =
\begin{cases}
\frac{1}{n}, & x = x_1, x_2, \dots, x_n \\
0, & \text{en otro caso.}
\end{cases}
$$

#### Propiedades

1. **No negatividad:**
   $$
   f_X(x) \ge 0
   $$

2. **Normalización:**
   $$
   \sum_{i=1}^{n} f_X(x_i) = 1
   $$

3. **Probabilidad de un evento:**
   Si $A \subseteq \{x_1, \dots, x_n\}$, entonces:
   $$
   P(X \in A) = \frac{|A|}{n}
   $$


#### Esperanza y Varianza

Cuando los valores son consecutivos, es decir:
$$
X \in \{a, a+1, a+2, \dots, b\}, \quad n = b - a + 1
$$

la **esperanza** y **varianza** están dadas por:

$$
\mathbb{E}[X] = \frac{a + b}{2}
$$

$$
\text{Var}(X) = \frac{(b - a + 1)^2 - 1}{12}
$$

#### Interpretación

La variable aleatoria uniforme discreta **modela situaciones de equiprobabilidad**.  
Cada resultado tiene la misma posibilidad de ocurrir, y la esperanza representa el **promedio teórico** de los valores posibles.

Gráficamente, la función de masa de probabilidad (f.m.p.) se representa como un conjunto de **barras de igual altura**, reflejando la equiprobabilidad de todos los valores de $X$.


### Bernoulli

#### Definicion

Una **variable aleatoria Bernoulli** modela un experimento aleatorio que solo puede tener **dos resultados posibles**: éxito o fracaso.  
Se usa en situaciones donde hay **una sola prueba** con un resultado binario, como lanzar una moneda o verificar si un componente funciona o no.

Sea $X$ una variable aleatoria que puede tomar los valores:

$$
X =
\begin{cases}
1, & \text{si ocurre el éxito}, \\
0, & \text{si ocurre el fracaso.}
\end{cases}
$$

Se dice que $X$ sigue una **distribución Bernoulli** con parámetro $p$ (probabilidad de éxito) si:

$$
P(X = 1) = p, \quad P(X = 0) = 1 - p, \quad 0 \le p \le 1
$$

Se denota:
$$
X \sim \text{Bernoulli}(p)
$$

#### Función de masa de probabilidad (f.m.p.)

La función de masa de probabilidad de una variable Bernoulli es:

$$
f_X(x) = p^x (1 - p)^{1 - x}, \quad x \in \{0,1\}
$$

De modo que:
- Si $x = 1$, entonces $f_X(1) = p$  
- Si $x = 0$, entonces $f_X(0) = 1 - p$


#### Propiedades

1. **No negatividad:**
   $$
   f_X(x) \ge 0, \quad \forall x \in \{0,1\}
   $$

2. **Normalización:**
   $$
   f_X(0) + f_X(1) = (1 - p) + p = 1
   $$

3. **Esperanza (valor esperado):**
   $$
   \mathbb{E}[X] = p
   $$

4. **Varianza:**
   $$
   \text{Var}(X) = p(1 - p)
   $$


#### Ejemplo — Lanzamiento de una moneda

Sea $X$ la variable aleatoria que vale 1 si sale **cara** y 0 si sale **cruz**.

- Si la moneda es justa, entonces $p = 0.5$.  
- Por tanto:
  $$
  f_X(0) = 0.5, \quad f_X(1) = 0.5
  $$
  $$
  \mathbb{E}[X] = 0.5, \quad \text{Var}(X) = 0.25
  $$

**Interpretación:**  
A largo plazo, el 50% de los lanzamientos resultarán en cara, y la varianza refleja la dispersión entre los resultados posibles (0 y 1).

La distribución Bernoulli se representa mediante dos barras verticales:
- En $x = 0$, la altura es $1 - p$ (fracaso),
- En $x = 1$, la altura es $p$ (éxito).


```python

## Variable Aleatoria Bernoulli
## Ejemplo: lanzamiento de una moneda.
## Visualización de la función de masa de probabilidad (f.m.p.),
## y cálculo de esperanza y varianza.

import numpy as np
import matplotlib.pyplot as plt

# Definición del parámetro p
p = 0.5  # probabilidad de éxito (por ejemplo, cara)
x = np.array([0, 1])  # valores posibles
f_x = np.array([1 - p, p])

# Esperanza y varianza
esperanza = p
varianza = p * (1 - p)

print(f"Parámetro p = {p}")
print(f"Esperanza (E[X]) = {esperanza:.2f}")
print(f"Varianza (Var[X]) = {varianza:.4f}")

# Gráfica de la f.m.p.
plt.figure(figsize=(5, 3))
plt.bar(x, f_x, color='darkorange', edgecolor='black', width=0.4)
plt.title(f"Distribución Bernoulli(p={p})", fontsize=12)
plt.xlabel("Valores de X")
plt.ylabel("f_X(x) = P(X = x)")
plt.xticks(x, ['0 (Fracaso)', '1 (Éxito)'])
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

### Binomial

#### Definicion

La **variable aleatoria binomial** modela el número de éxitos en un conjunto de $n$ ensayos independientes, donde cada ensayo tiene solo dos resultados posibles: éxito o fracaso.

Es una **extensión natural de la distribución Bernoulli**, que corresponde al caso $n = 1$.

Sea un experimento formado por $n$ ensayos independientes, donde en cada uno la probabilidad de éxito es $p$ y la de fracaso es $1 - p$.

Definimos la variable aleatoria:

$$
X = \text{número de éxitos en } n \text{ ensayos}
$$

Entonces $X$ sigue una **distribución binomial** con parámetros $n$ y $p$:

$$
X \sim \text{Binomial}(n, p)
$$

#### Función de masa de probabilidad (f.m.p.)

La función de probabilidad de $X$ es:

$$
f_X(x) = P(X = x) = \binom{n}{x} p^x (1 - p)^{n - x}, \quad x = 0, 1, 2, \dots, n
$$

donde:
- $\binom{n}{x} = \dfrac{n!}{x!(n-x)!}$ es el número de formas de obtener $x$ éxitos en $n$ ensayos.


#### Propiedades

1. **No negatividad:**
   $$
   f_X(x) \ge 0, \quad \forall x
   $$

2. **Normalización:**
   $$
   \sum_{x=0}^{n} f_X(x) = 1
   $$

3. **Esperanza:**
   $$
   \mathbb{E}[X] = n p
   $$

4. **Varianza:**
   $$
   \text{Var}(X) = n p (1 - p)
   $$

5. **Relación con Bernoulli:**
   Si $n = 1$, entonces:
   $$
   X \sim \text{Bernoulli}(p)
   $$


#### Ejemplo — Lanzamiento de una moneda tres veces

Sea $X$ el número de caras al lanzar una moneda **tres veces**.

Aquí:
$$
n = 3, \quad p = 0.5
$$

Entonces:

$$
f_X(x) = \binom{3}{x} (0.5)^x (0.5)^{3 - x}, \quad x = 0, 1, 2, 3
$$

Calculamos:
| $x$ | $f_X(x)$ |
|-----|-----------|
| 0 | $0.125$ |
| 1 | $0.375$ |
| 2 | $0.375$ |
| 3 | $0.125$ |

**Esperanza y varianza:**
$$
\mathbb{E}[X] = 3(0.5) = 1.5, \quad \text{Var}(X) = 3(0.5)(0.5) = 0.75
$$

**Interpretación:**  
Si se lanza una moneda tres veces, el número promedio de caras esperadas es 1.5, y la varianza mide la dispersión de los resultados posibles alrededor de ese valor.

La función de masa de probabilidad de una distribución binomial se representa con barras verticales, una por cada posible número de éxitos.

La forma de la distribución depende de los valores de $n$ y $p$:
- Si $p = 0.5$, es **simétrica**.
- Si $p < 0.5$ o $p > 0.5$, es **asimétrica**.

```python
## Variable Aleatoria Binomial

## Ejemplo: número de caras al lanzar una moneda tres veces.
## Visualización de la función de masa de probabilidad (f.m.p.),
## y cálculo de esperanza y varianza.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parámetros
n = 20       # número de ensayos
p = 0.5     # probabilidad de éxito
x = np.arange(0, n + 1)

# Función de masa de probabilidad
f_x = binom.pmf(x, n, p)

# Esperanza y varianza
esperanza = n * p
varianza = n * p * (1 - p)

print(f"n = {n}, p = {p}")
print(f"Esperanza (E[X]) = {esperanza:.2f}")
print(f"Varianza (Var[X]) = {varianza:.4f}")
print("Probabilidades:", np.round(f_x, 3))

# Gráfica de la f.m.p.
plt.figure(figsize=(6, 4))
plt.bar(x, f_x, color='royalblue', edgecolor='black')
plt.title(f"Distribución Binomial(n={n}, p={p})", fontsize=12)
plt.xlabel("Número de éxitos (x)")
plt.ylabel("f_X(x) = P(X = x)")
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

### Geométrica

#### Definición

La **variable aleatoria geométrica** modela el número de ensayos necesarios hasta que ocurre el **primer éxito** en una secuencia de ensayos de Bernoulli **independientes**, con probabilidad constante de éxito $p$.

Sea una secuencia de ensayos de Bernoulli independientes, cada uno con probabilidad de éxito $p$ y de fracaso $1 - p$.

La variable aleatoria:

$$
X = \text{número de ensayos necesarios hasta el primer éxito}
$$

sigue una **distribución geométrica** con parámetro $p$.

Se denota:

$$
X \sim \text{Geom}(p)
$$


#### Función de masa de probabilidad (f.m.p.)

La probabilidad de que el primer éxito ocurra en el ensayo número $x$ es:

$$
f_X(x) = P(X = x) = (1 - p)^{x - 1} p, \quad x = 1, 2, 3, \dots
$$

**Interpretación:**  
El evento “el primer éxito ocurre en el ensayo $x$” implica que los primeros $x - 1$ ensayos fueron fracasos y el $x$-ésimo fue éxito.


#### Propiedades

1. **Soporte:** $x = 1, 2, 3, \dots$

2. **No negatividad:**
   $$
   f_X(x) \ge 0
   $$

3. **Normalización:**
   $$
   \sum_{x=1}^{\infty} f_X(x) = \sum_{x=1}^{\infty} (1 - p)^{x-1} p = 1
   $$

4. **Esperanza:**
   $$
   \mathbb{E}[X] = \frac{1}{p}
   $$

5. **Varianza:**
   $$
   \text{Var}(X) = \frac{1 - p}{p^2}
   $$


#### Ejemplo - Lanzamiento de una moneda

Sea $X$ el número de lanzamientos necesarios para obtener la primera **cara** al lanzar una moneda equilibrada.

Aquí:
$$
p = 0.5, \quad X \sim \text{Geom}(0.5)
$$

La función de probabilidad es:

$$
f_X(x) = (1 - 0.5)^{x - 1}(0.5) = (0.5)^x, \quad x = 1, 2, 3, \dots
$$

**Ejemplos de probabilidades:**

| $x$ | $P(X = x)$ |
|-----|-------------|
| 1 | 0.5 |
| 2 | 0.25 |
| 3 | 0.125 |
| 4 | 0.0625 |

**Esperanza y varianza:**
$$
\mathbb{E}[X] = 2, \quad \text{Var}(X) = 2
$$

**Interpretación:**  
En promedio, se necesitan dos lanzamientos para obtener la primera cara, y la varianza mide la dispersión de esa espera.

La distribución geométrica tiene una forma **decreciente**, ya que las probabilidades disminuyen exponencialmente a medida que $x$ crece.  
Esto refleja que es más probable obtener un éxito temprano que uno muy tardío.


```python
## Variable Aleatoria Geométrica

## Ejemplo: número de lanzamientos necesarios para obtener la primera cara.
## Visualización de la función de masa de probabilidad (f.m.p.),
## y cálculo de esperanza y varianza.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

# Parámetro
p = 0.5  # probabilidad de éxito
x = np.arange(1, 11)  # primeros 10 valores posibles

# Función de masa de probabilidad
f_x = geom.pmf(x, p)

# Esperanza y varianza teóricas
esperanza = 1 / p
varianza = (1 - p) / (p**2)

print(f"Parámetro p = {p}")
print(f"Esperanza (E[X]) = {esperanza:.2f}")
print(f"Varianza (Var[X]) = {varianza:.2f}")
print("Probabilidades:", np.round(f_x, 3))

# Gráfica de la f.m.p.
plt.figure(figsize=(6, 4))
plt.bar(x, f_x, color='darkorange', edgecolor='black')
plt.title(f"Distribución Geométrica(p={p})", fontsize=12)
plt.xlabel("Número de ensayo hasta el primer éxito (x)")
plt.ylabel("f_X(x) = P(X = x)")
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

### Binomial Negativa

#### Definicion

La **variable aleatoria binomial negativa** modela el número de ensayos necesarios para obtener **r éxitos** en una secuencia de ensayos de Bernoulli independientes, donde la probabilidad de éxito en cada ensayo es constante e igual a $p$.

Sea una secuencia de ensayos de Bernoulli independientes, con probabilidad de éxito $p$ y de fracaso $1 - p$.

La variable aleatoria:

$$
X = \text{número total de ensayos necesarios hasta obtener } r \text{ éxitos}
$$

sigue una **distribución binomial negativa** con parámetros $r$ y $p$.

Se denota:

$$
X \sim \text{Binomial Negativa}(r, p)
$$


#### Función de masa de probabilidad (f.m.p.)

La probabilidad de que el $r$-ésimo éxito ocurra en el ensayo número $x$ es:

$$
f_X(x) = P(X = x) = \binom{x - 1}{r - 1} p^r (1 - p)^{x - r}, \quad x = r, r + 1, r + 2, \dots
$$

**Interpretación:**  
Para que el $r$-ésimo éxito ocurra en el ensayo $x$:
- Deben haberse producido $r - 1$ éxitos en los primeros $x - 1$ ensayos,  
- Y el ensayo número $x$ debe ser un éxito.


#### Propiedades

1. **Soporte:** $x = r, r + 1, r + 2, \dots$

2. **No negatividad:**
   $$
   f_X(x) \ge 0
   $$

3. **Normalización:**
   $$
   \sum_{x = r}^{\infty} f_X(x) = 1
   $$

4. **Esperanza:**
   $$
   \mathbb{E}[X] = \frac{r}{p}
   $$

5. **Varianza:**
   $$
   \text{Var}(X) = \frac{r(1 - p)}{p^2}
   $$

6. **Relación con la geométrica:**
   Si $r = 1$, entonces:
   $$
   X \sim \text{Geom}(p)
   $$


#### Ejemplo — Pozos petroleros

Una compañía petrolera perfora pozos con probabilidad $p = 0.2$ de encontrar petróleo.  
Sea $X$ el número de pozos que deben perforarse para obtener **r = 3** pozos productores.

Entonces:
$$
X \sim \text{Binomial Negativa}(r = 3, p = 0.2)
$$

La función de probabilidad es:

$$
f_X(x) = \binom{x - 1}{2}(0.2)^3(0.8)^{x - 3}, \quad x = 3, 4, 5, \dots
$$

**Ejemplo numérico:**

- Probabilidad de obtener el tercer pozo productor en el séptimo pozo perforado:

$$
P(X = 7) = \binom{6}{2}(0.2)^3(0.8)^4 = 15(0.008)(0.4096) = 0.04915
$$

**Esperanza y Varianza:**
$$
\mathbb{E}[X] = \frac{3}{0.2} = 15, \quad \text{Var}(X) = \frac{3(0.8)}{0.04} = 60
$$

**Interpretación:**  
En promedio, deben perforarse 15 pozos para obtener tres productores, pero la alta varianza refleja la gran incertidumbre inherente al proceso.

La distribución binomial negativa tiene una forma **decreciente** y **asimétrica**, concentrando mayor probabilidad en los valores más bajos de $x$.  
A medida que aumenta $p$, los éxitos se logran antes y la distribución se concentra a la izquierda.  
Si $p$ es bajo, la distribución se dispersa más a la derecha.


```python

## Variable Aleatoria Binomial Negativa

## Ejemplo: número de pozos que deben perforarse hasta obtener r pozos productores.
## Visualización de la función de masa de probabilidad (f.m.p.),
## y cálculo de esperanza y varianza.


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom

# Parámetros
r = 3       # número de éxitos requeridos
p = 0.2     # probabilidad de éxito
x = np.arange(r, 40)  # valores posibles (desde r en adelante)

# Función de masa de probabilidad
f_x = nbinom.pmf(x - r, r, p)  # scipy define X como # de fracasos antes del r-ésimo éxito

# Esperanza y varianza teóricas
esperanza = r / p
varianza = r * (1 - p) / (p**2)

print(f"r = {r}, p = {p}")
print(f"Esperanza (E[X]) = {esperanza:.2f}")
print(f"Varianza (Var[X]) = {varianza:.2f}")
print("Probabilidades (primeros valores):", np.round(f_x[:10], 4))

# Gráfica de la f.m.p.
plt.figure(figsize=(7, 4))
plt.bar(x, f_x, color='seagreen', edgecolor='black')
plt.title(f"Distribución Binomial Negativa (r={r}, p={p})", fontsize=12)
plt.xlabel("Número total de ensayos hasta el r-ésimo éxito (x)")
plt.ylabel("f_X(x) = P(X = x)")
plt.xticks(np.arange(r, 40, 2))
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

### Poisson

#### Definicion

La **variable aleatoria Poisson** modela el número de veces que ocurre un evento en un intervalo de tiempo o espacio, **bajo las siguientes condiciones**:

1. Los eventos ocurren **de manera independiente** unos de otros.  
2. La **tasa promedio de ocurrencia** es constante y conocida.  
3. Dos eventos no pueden ocurrir exactamente al mismo tiempo (son mutuamente excluyentes en un instante).

Sea $X$ una variable aleatoria que representa el número de eventos ocurridos en un intervalo dado.  
Se dice que $X$ sigue una **distribución de Poisson** con parámetro $\lambda > 0$ si:

$$
P(X = x) = f_X(x) = \frac{e^{-\lambda} \lambda^x}{x!}, \quad x = 0, 1, 2, \dots
$$

donde:
- $\lambda$ es el **número promedio de eventos** que ocurren en el intervalo.  
- $e \approx 2.71828$ es la base de los logaritmos naturales.

Se denota:

$$
X \sim \text{Poisson}(\lambda)
$$


#### Propiedades

1. **Soporte:** $x = 0, 1, 2, \dots$

2. **No negatividad:**
   $$
   f_X(x) \ge 0
   $$

3. **Normalización:**
   $$
   \sum_{x=0}^{\infty} f_X(x) = e^{-\lambda} \sum_{x=0}^{\infty} \frac{\lambda^x}{x!} = e^{-\lambda} e^{\lambda} = 1
   $$

4. **Esperanza:**
   $$
   \mathbb{E}[X] = \lambda
   $$

5. **Varianza:**
   $$
   \text{Var}(X) = \lambda
   $$

6. **Propiedad de aditividad:**
   Si $X_1 \sim \text{Poisson}(\lambda_1)$ y $X_2 \sim \text{Poisson}(\lambda_2)$ son independientes, entonces:
   $$
   X_1 + X_2 \sim \text{Poisson}(\lambda_1 + \lambda_2)
   $$

#### Interpretación

El parámetro $\lambda$ representa tanto la **media** como la **varianza** de la distribución.  
Esto significa que la dispersión de los datos crece junto con la tasa promedio de ocurrencia.


#### Ejemplo — Llegadas a una gasolinera

Supongamos que los autos llegan a una gasolinera a una tasa promedio de $\lambda = 3$ autos por minuto.

Sea $X$ el número de autos que llegan en un minuto.

$$
X \sim \text{Poisson}(3)
$$

**Probabilidad de que lleguen exactamente 2 autos:**

$$
P(X = 2) = \frac{e^{-3} 3^2}{2!} = e^{-3} \cdot \frac{9}{2} = 0.2240
$$

**Esperanza y varianza:**
$$
\mathbb{E}[X] = \lambda = 3, \quad \text{Var}(X) = 3
$$

**Interpretación:**  
En promedio llegan tres autos por minuto, y la probabilidad de que lleguen exactamente dos es aproximadamente 22.4%.

La función de masa de probabilidad de una Poisson presenta una **forma sesgada a la derecha** cuando $\lambda$ es pequeño,  
y se vuelve **más simétrica** conforme $\lambda$ crece.


```python
## Variable Aleatoria Poisson

## Ejemplo: número de autos que llegan a una gasolinera por minuto.
## Visualización de la función de masa de probabilidad (f.m.p.),
## y cálculo de esperanza y varianza.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parámetro
lam = 3  # tasa promedio de ocurrencia
x = np.arange(0, 13)  # valores posibles (0 a 12 eventos)

# Función de masa de probabilidad
f_x = poisson.pmf(x, lam)

# Esperanza y varianza teóricas
esperanza = lam
varianza = lam

print(f"λ = {lam}")
print(f"Esperanza (E[X]) = {esperanza:.2f}")
print(f"Varianza (Var[X]) = {varianza:.2f}")
print("Probabilidades:", np.round(f_x, 3))

# Gráfica de la f.m.p.
plt.figure(figsize=(7, 4))
plt.bar(x, f_x, color='royalblue', edgecolor='black')
plt.title(f"Distribución Poisson(λ={lam})", fontsize=12)
plt.xlabel("Número de eventos (x)")
plt.ylabel("f_X(x) = P(X = x)")
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

### Hipergeométrica

#### Definicion

Sea una población de tamaño $N$ que contiene:
- $K$ elementos clasificados como **éxitos**,
- $N - K$ elementos clasificados como **fracasos**.

Se extrae una muestra **sin reemplazo** de tamaño $n$.

Definimos la variable aleatoria:

$$
X = \text{número de éxitos obtenidos en la muestra de tamaño } n
$$

Entonces $X$ sigue una **distribución hipergeométrica** con parámetros $N$, $K$ y $n$.

Se denota:

$$
X \sim \text{Hipergeométrica}(N, K, n)
$$

#### Función de masa de probabilidad (f.m.p.)

La probabilidad de obtener exactamente $x$ éxitos en la muestra es:

$$
f_X(x) = P(X = x) = 
\frac{\displaystyle\binom{K}{x}\binom{N - K}{n - x}}{\displaystyle\binom{N}{n}},
\quad \max(0, n - (N - K)) \le x \le \min(n, K)
$$

**Interpretación:**  
- $\binom{K}{x}$: formas de elegir $x$ éxitos de los $K$ disponibles.  
- $\binom{N - K}{n - x}$: formas de elegir los demás $(n - x)$ fracasos.  
- $\binom{N}{n}$: formas totales de elegir la muestra.


#### Propiedades

1. **Esperanza:**
   $$
   \mathbb{E}[X] = n \frac{K}{N}
   $$

2. **Varianza:**
   $$
   \text{Var}(X) = n \frac{K}{N}\left(1 - \frac{K}{N}\right)\frac{N - n}{N - 1}
   $$

3. **Relación con la binomial:**
   Si la población es muy grande comparada con la muestra ($N \gg n$),  
   entonces la hipergeométrica se aproxima a una **binomial** $B(n, p)$, con $p = K/N$.


#### Ejemplo — Control de calidad

Una fábrica produce $N = 50$ piezas, de las cuales $K = 10$ son defectuosas.  
Se selecciona una muestra de $n = 5$ piezas **sin reemplazo**.

Sea $X$ el número de piezas defectuosas en la muestra.

Entonces:
$$
X \sim \text{Hipergeométrica}(50, 10, 5)
$$

**Probabilidad de obtener exactamente 2 defectuosas:**

$$
P(X = 2) = \frac{\binom{10}{2}\binom{40}{3}}{\binom{50}{5}}
$$

Calculando:
$$
P(X = 2) = \frac{45 \cdot 9880}{2{,}118{,}760} = 0.21
$$

**Esperanza y Varianza:**
$$
\mathbb{E}[X] = 5\left(\frac{10}{50}\right) = 1, \quad
\text{Var}(X) = 5\left(\frac{10}{50}\right)\left(1 - \frac{10}{50}\right)\frac{45}{49} = 0.918
$$

**Interpretación:**  
En promedio se espera encontrar **1 pieza defectuosa** por muestra,  
con una varianza que refleja la dispersión debido al muestreo sin reemplazo.

La distribución hipergeométrica puede ser **asimétrica** o casi **simétrica** dependiendo de las proporciones $K/N$ y del tamaño de muestra $n$.

Si $p = K/N$ es pequeño, la distribución se concentra cerca de 0 (pocos éxitos esperados).  
Si $p$ es grande, se concentra hacia valores altos de $x$.


```python
## Variable Aleatoria Hipergeométrica

## Ejemplo: control de calidad — número de piezas defectuosas en una muestra sin reemplazo.
## Visualización de la función de masa de probabilidad (f.m.p.),
## y cálculo de esperanza y varianza.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import hypergeom

# Parámetros
N = 50  # tamaño de la población
K = 10  # número de éxitos en la población (defectuosas)
n = 5   # tamaño de la muestra
x = np.arange(0, n + 1)  # posibles valores de éxitos en la muestra

# Función de masa de probabilidad
f_x = hypergeom.pmf(x, N, K, n)

# Esperanza y varianza teóricas
esperanza = n * (K / N)
varianza = n * (K / N) * (1 - K / N) * ((N - n) / (N - 1))

print(f"N = {N}, K = {K}, n = {n}")
print(f"Esperanza (E[X]) = {esperanza:.3f}")
print(f"Varianza (Var[X]) = {varianza:.3f}")
print("Probabilidades:", np.round(f_x, 3))

# Gráfica de la f.m.p.
plt.figure(figsize=(6, 4))
plt.bar(x, f_x, color='darkorange', edgecolor='black')
plt.title(f"Distribución Hipergeométrica(N={N}, K={K}, n={n})", fontsize=12)
plt.xlabel("Número de éxitos en la muestra (x)")
plt.ylabel("f_X(x) = P(X = x)")
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

## Variables Aleatorias Continuas

### Uniforme Continua

#### Definicion

La **variable aleatoria uniforme continua** describe fenómenos en los que **todos los valores dentro de un intervalo** tienen la **misma probabilidad** de ocurrir.  
Es una de las distribuciones continuas más simples y fundamentales.

Sea $X$ una variable aleatoria continua tal que:

$$
X \sim \text{Uniforme}(a, b)
$$

si su función de densidad de probabilidad (f.d.p.) está dada por:

$$
f_X(x) =
\begin{cases}
\dfrac{1}{b - a}, & a \le x \le b, \\
0, & \text{en otro caso.}
\end{cases}
$$

donde:
- $a$ y $b$ son los **límites del intervalo**, con $a < b$.
- La probabilidad está **uniformemente distribuida** entre $a$ y $b$.


#### Propiedades

1. **Soporte:** $x \in [a, b]$

2. **No negatividad:**
   $$
   f_X(x) \ge 0, \quad \forall x
   $$

3. **Normalización:**
   $$
   \int_a^b f_X(x)\,dx = \int_a^b \frac{1}{b - a}\,dx = 1
   $$

4. **Función de distribución acumulada (F.D.A.):**
   $$
   F_X(x) =
   \begin{cases}
   0, & x < a \\
   \dfrac{x - a}{b - a}, & a \le x \le b \\
   1, & x > b
   \end{cases}
   $$

5. **Esperanza:**
   $$
   \mathbb{E}[X] = \frac{a + b}{2}
   $$

6. **Varianza:**
   $$
   \text{Var}(X) = \frac{(b - a)^2}{12}
   $$


#### Ejemplo — Tiempo de espera en una parada de autobús

Supongamos que el tiempo de espera $X$ (en minutos) para la llegada del autobús se distribuye uniformemente entre 5 y 15 minutos.

$$
X \sim \text{Uniforme}(5, 15)
$$

Entonces:

- La **densidad** es constante:  
  $$
  f_X(x) = \frac{1}{15 - 5} = 0.1
  $$

- La **probabilidad de esperar entre 8 y 12 minutos** es:
  $$
  P(8 \le X \le 12) = \int_8^{12} 0.1\,dx = (12 - 8)(0.1) = 0.4
  $$

- La **esperanza y varianza** son:
  $$
  \mathbb{E}[X] = 10, \quad \text{Var}(X) = \frac{(10)^2}{12} = 8.33
  $$

**Interpretación:**  
El tiempo promedio de espera es 10 minutos, y la dispersión es moderada ya que todos los tiempos son igualmente probables dentro del rango.


```python
## Variable Aleatoria Uniforme Continua

## Ejemplo: tiempo de espera en una parada de autobús entre 5 y 15 minutos.
## Visualización de la función de densidad (f.d.p.),
## cálculo de probabilidad, esperanza y varianza.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Parámetros
a = 5   # límite inferior
b = 15  # límite superior
x = np.linspace(a - 2, b + 2, 300)  # rango extendido para graficar

# Función de densidad de probabilidad
f_x = uniform.pdf(x, loc=a, scale=b - a)

# Cálculos teóricos
esperanza = (a + b) / 2
varianza = ((b - a) ** 2) / 12

# Probabilidad de esperar entre 8 y 12 minutos
p_8_12 = uniform.cdf(12, loc=a, scale=b - a) - uniform.cdf(8, loc=a, scale=b - a)

print(f"a = {a}, b = {b}")
print(f"Esperanza (E[X]) = {esperanza:.2f}")
print(f"Varianza (Var[X]) = {varianza:.2f}")
print(f"P(8 ≤ X ≤ 12) = {p_8_12:.2f}")

# Gráfica de la f.d.p.
plt.figure(figsize=(7, 4))
plt.plot(x, f_x, color='darkorange', lw=2)
plt.title(f"Distribución Uniforme Continua({a}, {b})", fontsize=12)
plt.xlabel("x (tiempo en minutos)")
plt.ylabel("f_X(x) = 1 / (b - a)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.fill_between(x, 0, f_x, where=(x >= 8) & (x <= 12), color='orange', alpha=0.4, label='P(8 ≤ X ≤ 12)')
plt.legend()
plt.show()
```

### Exponencial

#### Definición

La **variable aleatoria exponencial** modela el **tiempo entre la ocurrencia de dos eventos consecutivos** en un proceso que ocurre de forma continua y con una tasa constante.

Se utiliza ampliamente en:
- Modelos de **tiempos de vida** (por ejemplo, de componentes o equipos),
- **Tiempos de espera** en colas,
- Procesos **Poisson** (donde el número de eventos por intervalo sigue una Poisson y el tiempo entre ellos una exponencial).

Sea $X$ una variable aleatoria continua que toma valores positivos.  
Se dice que $X$ sigue una **distribución exponencial** con parámetro $\lambda > 0$ si su función de densidad de probabilidad (f.d.p.) está dada por:

$$
f_X(x) =
\begin{cases}
\lambda e^{-\lambda x}, & x \ge 0, \\
0, & x < 0.
\end{cases}
$$

Se denota:
$$
X \sim \text{Exp}(\lambda)
$$

donde:
- $\lambda$ es la **tasa de ocurrencia** o **intensidad** del proceso,
- $\frac{1}{\lambda}$ es la **media** o **tiempo promedio entre eventos**.


#### Propiedades

1. **Soporte:** $x \ge 0$

2. **No negatividad:**
   $$
   f_X(x) \ge 0
   $$

3. **Normalización:**
   $$
   \int_0^{\infty} \lambda e^{-\lambda x} dx = 1
   $$

4. **Función de distribución acumulada (F.D.A.):**
   $$
   F_X(x) =
   \begin{cases}
   0, & x < 0 \\
   1 - e^{-\lambda x}, & x \ge 0
   \end{cases}
   $$

5. **Esperanza:**
   $$
   \mathbb{E}[X] = \frac{1}{\lambda}
   $$

6. **Varianza:**
   $$
   \text{Var}(X) = \frac{1}{\lambda^2}
   $$

7. **Propiedad de falta de memoria:**
   $$
   P(X > s + t \mid X > s) = P(X > t)
   $$
   Es decir, el tiempo adicional hasta el siguiente evento **no depende del tiempo ya transcurrido**.


#### Ejemplo — Tiempo de atención en una ventanilla

Supongamos que el tiempo de atención de un cliente ($X$) en una ventanilla se distribuye exponencialmente con una tasa $\lambda = 0.5$ clientes por minuto.

Esto significa que, en promedio, el empleado atiende a:
$$
\frac{1}{\lambda} = 2 \text{ minutos por cliente.}
$$

**a)** Probabilidad de que el tiempo de atención sea menor a 3 minutos:

$$
P(X < 3) = 1 - e^{-\lambda(3)} = 1 - e^{-1.5} = 0.7769
$$

**b)** Esperanza y varianza:
$$
\mathbb{E}[X] = 2, \quad \text{Var}(X) = 4
$$

**Interpretación:**  
Existe una probabilidad del **77.7%** de que el empleado atienda al cliente en menos de 3 minutos, con un tiempo promedio de atención de 2 minutos.

La función de densidad exponencial tiene una forma **decreciente**, partiendo de su valor máximo en $x = 0$.  
A medida que $\lambda$ aumenta, los eventos ocurren más frecuentemente (tiempos más cortos entre eventos).  
Si $\lambda$ disminuye, los tiempos de espera se alargan.

```python
## Variable Aleatoria Exponencial
## Ejemplo: tiempo de atención en una ventanilla con tasa λ = 0.5 clientes por minuto.
## Visualización de la función de densidad (f.d.p.),
## cálculo de probabilidad, esperanza y varianza.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Parámetro
lam = 0.5  # tasa (clientes por minuto)
x = np.linspace(0, 12, 300)  # rango de valores

# Función de densidad de probabilidad
f_x = expon.pdf(x, scale=1/lam)

# Cálculos teóricos
esperanza = 1 / lam
varianza = 1 / (lam ** 2)

# Probabilidad de ser atendido en menos de 3 minutos
p_menor_3 = expon.cdf(3, scale=1/lam)

print(f"λ = {lam}")
print(f"Esperanza (E[X]) = {esperanza:.2f}")
print(f"Varianza (Var[X]) = {varianza:.2f}")
print(f"P(X < 3) = {p_menor_3:.4f}")

# Gráfica de la f.d.p.
plt.figure(figsize=(7, 4))
plt.plot(x, f_x, color='seagreen', lw=2)
plt.title(f"Distribución Exponencial(λ={lam})", fontsize=12)
plt.xlabel("x (tiempo de atención en minutos)")
plt.ylabel("f_X(x) = λ e^(-λx)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.fill_between(x, 0, f_x, where=(x <= 3), color='lightblue', alpha=0.4, label='P(X < 3)')
plt.legend()
plt.show()
```

### Normal (o Gaussiana)

#### Definicion

La **variable aleatoria normal** modela fenómenos naturales y sociales que resultan de la **suma de múltiples efectos pequeños e independientes**.  
Es la distribución continua más utilizada en estadística y probabilidad.

Sea $X$ una variable aleatoria continua con media $\mu$ y desviación estándar $\sigma > 0$.

Se dice que $X$ sigue una **distribución normal** con parámetros $\mu$ y $\sigma$ si su función de densidad de probabilidad (f.d.p.) es:

$$
f_X(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \, e^{ -\frac{(x - \mu)^2}{2\sigma^2} }, \quad x \in \mathbb{R}
$$

Se denota:
$$
X \sim N(\mu, \sigma^2)
$$

donde:
- $\mu$ representa la **media** o **centro** de la distribución,
- $\sigma$ controla la **dispersión** (qué tan concentrados o extendidos están los valores).


#### Propiedades

1. **Simetría:**  
   La distribución es perfectamente **simétrica** respecto a $\mu$.

2. **Esperanza y Varianza:**
   $$
   \mathbb{E}[X] = \mu, \quad \text{Var}(X) = \sigma^2
   $$

3. **Probabilidades aproximadas (Regla Empírica 68–95–99.7):**
   - 68% de los valores están dentro de $[\mu - \sigma, \mu + \sigma]$
   - 95% dentro de $[\mu - 2\sigma, \mu + 2\sigma]$
   - 99.7% dentro de $[\mu - 3\sigma, \mu + 3\sigma]$

4. **Estándarización:**  
   Si $X \sim N(\mu, \sigma^2)$, entonces:
   $$
   Z = \frac{X - \mu}{\sigma} \sim N(0, 1)
   $$
   donde $Z$ se llama **variable normal estándar**.

#### Ejemplo — Alturas de estudiantes universitarios

Supongamos que las alturas (en cm) de los estudiantes universitarios se distribuyen normalmente con media $\mu = 170$ cm y desviación estándar $\sigma = 6$ cm.

$$
X \sim N(170, 6^2)
$$

**a)** Probabilidad de que un estudiante mida menos de 175 cm:

$$
P(X < 175) = P\left(Z < \frac{175 - 170}{6}\right) = P(Z < 0.8333) = 0.7977
$$

**b)** Probabilidad de que mida entre 160 y 180 cm:

$$
P(160 < X < 180) = P(-1.67 < Z < 1.67) = 0.9050
$$

**Esperanza y Varianza:**
$$
\mathbb{E}[X] = 170, \quad \text{Var}(X) = 36
$$

**Interpretación:**  
La mayoría de los estudiantes tienen alturas entre 160 y 180 cm.  
Solo valores extremos se alejan de la media más de 2 desviaciones estándar.

La **curva normal** tiene forma de campana:
- Su punto máximo está en $x = \mu$,
- Es simétrica,
- Se aproxima a cero en los extremos,
- Cuanto mayor sea $\sigma$, más **ancha y baja** será la curva.


```python
## Variable Aleatoria Normal (o Gaussiana)

## Ejemplo: alturas de estudiantes con μ = 170 cm y σ = 6 cm.
## Visualización de la función de densidad (f.d.p.),
## cálculo de probabilidades, esperanza y varianza.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros
mu = 170  # media
sigma = 6  # desviación estándar
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 400)  # rango de valores

# Función de densidad de probabilidad
f_x = norm.pdf(x, mu, sigma)

# Cálculos teóricos
esperanza = mu
varianza = sigma ** 2

# Probabilidades
p_menor_175 = norm.cdf(175, mu, sigma)
p_entre_160_180 = norm.cdf(180, mu, sigma) - norm.cdf(160, mu, sigma)

print(f"μ = {mu}, σ = {sigma}")
print(f"Esperanza (E[X]) = {esperanza:.2f}")
print(f"Varianza (Var[X]) = {varianza:.2f}")
print(f"P(X < 175) = {p_menor_175:.4f}")
print(f"P(160 < X < 180) = {p_entre_160_180:.4f}")

# Gráfica de la f.d.p.
plt.figure(figsize=(7, 4))
plt.plot(x, f_x, color='royalblue', lw=2, label='f_X(x)')
plt.fill_between(x, 0, f_x, where=(x >= 160) & (x <= 180), color='skyblue', alpha=0.4, label='P(160 < X < 180)')
plt.title(f"Distribución Normal(μ={mu}, σ={sigma})", fontsize=12)
plt.xlabel("Altura (cm)")
plt.ylabel("f_X(x)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### Gamma

#### Definicion

La **variable aleatoria Gamma** es una de las distribuciones continuas más importantes.  
Generaliza la distribución exponencial y se usa para modelar **tiempos de espera acumulados** hasta que ocurren múltiples eventos en un proceso Poisson.

Sea $X$ una variable aleatoria continua con parámetros:
- $α > 0$: parámetro de **forma**,
- $β > 0$: parámetro de **escala** (a veces se usa $\lambda = 1/β$ como tasa).

Se dice que $X$ sigue una **distribución Gamma** con parámetros $α$ y $β$ si su función de densidad de probabilidad (f.d.p.) es:

$$
f_X(x) =
\begin{cases}
\dfrac{1}{\Gamma(α) β^α} x^{α - 1} e^{-x/β}, & x > 0, \\
0, & x \le 0,
\end{cases}
$$

donde $\Gamma(α)$ es la **función Gamma**, definida como:

$$
\Gamma(α) = \int_0^{\infty} t^{α - 1} e^{-t} dt
$$

#### Propiedades

1. **Soporte:** $x > 0$

2. **Esperanza y Varianza:**
   $$
   \mathbb{E}[X] = αβ, \quad \text{Var}(X) = αβ^2
   $$

3. **Función de distribución acumulada (F.D.A.):**
   $$
   F_X(x) = \frac{1}{\Gamma(α)} \gamma\left(α, \frac{x}{β}\right)
   $$
   donde $\gamma(α, x)$ es la **función Gamma incompleta**.

4. **Relaciones importantes:**
   - Si $α = 1$, la distribución Gamma se reduce a una **exponencial** con parámetro $\lambda = 1/β$.
   - Si $α = k/2$ y $β = 2$, se obtiene la **distribución chi-cuadrado** con $k$ grados de libertad.


#### Interpretación

La variable Gamma puede entenderse como el **tiempo total hasta que ocurren $α$ eventos** en un proceso Poisson con tasa $1/β$.

Por ejemplo:
- Si $α = 1$: tiempo hasta el **primer** evento → exponencial.
- Si $α = 3$: tiempo hasta el **tercer** evento → gamma.

#### Ejemplo — Vida útil de un sistema

Supongamos que el tiempo de vida útil $X$ (en horas) de un sistema depende de la duración de **tres componentes independientes**, cada una con una vida media de $β = 100$ horas.  
Entonces:
$$
X \sim \text{Gamma}(α = 3, β = 100)
$$

**a)** Esperanza y Varianza:
$$
\mathbb{E}[X] = αβ = 300, \quad \text{Var}(X) = αβ^2 = 30{,}000
$$

**b)** Probabilidad de que el sistema dure menos de 250 horas:
$$
P(X < 250) = F_X(250)
$$

**Interpretación:**  
En promedio, el sistema dura **300 horas**, pero hay una variabilidad importante (varianza 30 000).  
La probabilidad de que falle antes de las 250 horas puede calcularse con la función de distribución acumulada.

- Para valores pequeños de $α$, la distribución es **asimétrica a la derecha**.  
- Conforme $α$ aumenta, la curva se vuelve más **simétrica** y se aproxima a una forma normal.  
- El parámetro $β$ **escala** la distribución en el eje horizontal.


```python
## Variable Aleatoria Gamma

## Ejemplo: vida útil de un sistema formado por 3 componentes con β = 100 horas.
## Visualización de la función de densidad (f.d.p.),
## cálculo de probabilidad, esperanza y varianza.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Parámetros
alpha = 3   # parámetro de forma
beta = 100  # parámetro de escala
x = np.linspace(0, 800, 400)  # rango de valores

# Función de densidad de probabilidad
f_x = gamma.pdf(x, a=alpha, scale=beta)

# Cálculos teóricos
esperanza = alpha * beta
varianza = alpha * (beta ** 2)

# Probabilidad de que el sistema dure menos de 250 horas
p_menor_250 = gamma.cdf(250, a=alpha, scale=beta)

print(f"α = {alpha}, β = {beta}")
print(f"Esperanza (E[X]) = {esperanza:.2f}")
print(f"Varianza (Var[X]) = {varianza:.2f}")
print(f"P(X < 250) = {p_menor_250:.4f}")

# Gráfica de la f.d.p.
plt.figure(figsize=(7, 4))
plt.plot(x, f_x, color='darkorange', lw=2)
plt.title(f"Distribución Gamma(α={alpha}, β={beta})", fontsize=12)
plt.xlabel("x (tiempo de vida del sistema en horas)")
plt.ylabel("f_X(x)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.fill_between(x, 0, f_x, where=(x <= 250), color='orange', alpha=0.4, label='P(X < 250)')
plt.legend()
plt.show()
```

## Simulación de Variables Aleatorias

### Método de la Transformada Inversa

#### Definicion

El **método de la transformada inversa** permite generar valores de una variable aleatoria $X$ con una **distribución de probabilidad conocida** (a través de su función de distribución acumulada), utilizando una **variable aleatoria uniforme** $U \sim U(0, 1)$.


#### Idea general

Si $X$ tiene función de distribución acumulada (F.D.A.) $F_X(x)$, entonces:

$$
F_X(x) = P(X \le x)
$$

Si $U \sim U(0,1)$, entonces $U = F_X(X)$ tiene una distribución uniforme.  
Por lo tanto, si invertimos esta relación:

$$
X = F_X^{-1}(U)
$$

obtenemos una variable aleatoria $X$ con la misma distribución que la original.


#### Pasos del método

1. Generar una variable aleatoria uniforme $U \sim U(0, 1)$.
2. Calcular el valor de $X = F_X^{-1}(U)$, donde $F_X^{-1}$ es la **función inversa de la distribución acumulada**.
3. Repetir el proceso para obtener tantas observaciones de $X$ como se desee.


#### Ventajas

- Es **exacto**, siempre que $F_X^{-1}$ pueda calcularse de forma analítica.  
- Sencillo de implementar.  
- Permite generar valores de muchas distribuciones (uniforme, exponencial, Cauchy, etc.).

**Limitación:**  
Cuando $F_X^{-1}$ no tiene una forma cerrada (por ejemplo, en la normal), se requieren métodos alternativos como **aceptación-rechazo** o **Box–Muller**.


#### Ejemplo — Simulación de una variable exponencial

Supongamos que queremos simular una variable $X \sim \text{Exp}(\lambda)$, cuya F.D.A. es:

$$
F_X(x) = 1 - e^{-\lambda x}
$$

Entonces, la inversa es:

$$
F_X^{-1}(u) = -\frac{1}{\lambda} \ln(1 - u)
$$

Si $U \sim U(0,1)$, generamos:
$$
X = -\frac{1}{\lambda} \ln(1 - U)
$$

y $X$ tendrá distribución exponencial con parámetro $\lambda$.

##### Ejemplo numérico

Sea $\lambda = 2$.

1. Generamos $U = 0.7$  
2. Calculamos:  
   $$
   X = -\frac{1}{2}\ln(1 - 0.7) = 0.6019
   $$
3. Este valor representa una observación simulada de $X \sim \text{Exp}(2)$.

Al simular muchas observaciones y graficar su histograma, obtenemos la forma característica de la función de densidad exponencial.


```python
## Simulación de Variables Aleatorias — Transformada Inversa
## Ejemplo: Generación de valores de una variable exponencial X ~ Exp(λ)
## Comparación entre histograma simulado y densidad teórica.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Parámetro
lam = 2  # tasa λ
n = 100000  # número de simulaciones

# 1. Generar variables uniformes U(0,1)
U = np.random.rand(n)

# 2. Aplicar la transformada inversa: X = -1/λ * ln(1 - U)
X = - (1 / lam) * np.log(1 - U)

# 3. Comparación teórica
x_vals = np.linspace(0, 3, 200)
f_teorica = expon.pdf(x_vals, scale=1/lam)

# Esperanza y varianza teóricas
esperanza = 1 / lam
varianza = 1 / (lam ** 2)

print(f"λ = {lam}")
print(f"Esperanza teórica = {esperanza:.3f}")
print(f"Varianza teórica = {varianza:.3f}")
print(f"Media simulada = {np.mean(X):.3f}")
print(f"Varianza simulada = {np.var(X):.3f}")

# Gráfica
plt.figure(figsize=(7, 4))
plt.hist(X, bins=50, density=True, color='lightgreen', edgecolor='black', alpha=0.6, label='Simulación')
plt.plot(x_vals, f_teorica, color='darkgreen', lw=2, label='f.d.p. teórica')
plt.title(f"Simulación por Transformada Inversa — Exp(λ={lam})", fontsize=12)
plt.xlabel("x")
plt.ylabel("Densidad")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

## Función Generadora de Momentos (FGM)

#### Definicion

Sea $X$ una variable aleatoria con esperanza finita.  
Su **función generadora de momentos (FGM)** se define como:

$$
M_X(t) = \mathbb{E}[e^{tX}], \quad t \in \mathbb{R}
$$

#### Propiedades

1. $M_X(0) = 1$  
2. Si $M_X(t)$ existe en un entorno de 0, **determina completamente la distribución** de $X$.  
3. Los **momentos** se obtienen derivando $M_X(t)$ en $t = 0$:
   $$
   \mathbb{E}[X^r] = M_X^{(r)}(0)
   $$

#### Ejemplo

Si $X \sim N(\mu, \sigma^2)$, entonces:
$$
M_X(t) = e^{\mu t + \frac{1}{2}\sigma^2 t^2}
$$

De donde:
- $\mathbb{E}[X] = M_X'(0) = \mu$  
- $\text{Var}(X) = M_X''(0) - (M_X'(0))^2 = \sigma^2$

```python
## Ejemplo: Función generadora de momentos de una variable normal

import sympy as sp

# Definición simbólica
t, mu, sigma = sp.symbols('t mu sigma', real=True)
M = sp.exp(mu * t + (sigma**2 * t**2) / 2)

# Derivadas para los primeros momentos
M1 = sp.diff(M, t).subs(t, 0)
M2 = sp.diff(M, t, 2).subs(t, 0)
Var = M2 - M1**2

print(f"M_X(t) = {sp.simplify(M)}")
print(f"E[X] = {M1}")
print(f"Var[X] = {Var}")
```

## Función Característica (FC)

#### Definición

Sea $X$ una variable aleatoria.  
Su **función característica** se define como:

$$
\varphi_X(t) = \mathbb{E}[e^{itX}], \quad t \in \mathbb{R}
$$

donde $i = \sqrt{-1}$.

#### Propiedades

1. $\varphi_X(0) = 1$
2. $|\varphi_X(t)| \le 1$
3. Si $\varphi_X(t)$ existe para todo $t$, **determina completamente la distribución**.
4. Los momentos se obtienen como:
   $$
   \mathbb{E}[X^r] = \frac{1}{i^r} \varphi_X^{(r)}(0)
   $$


#### Ejemplo

Si $X \sim N(0, 1)$:

$$
\varphi_X(t) = e^{-\frac{t^2}{2}}
$$

De donde:
- $\mathbb{E}[X] = 0$
- $\text{Var}(X) = 1$

**Interpretación:**  
La función característica es una versión compleja y más general que la FGM.  
Siempre existe, incluso cuando la FGM no converge.

## Ley de los Grandes Números (LGN)

#### Enunciado

Sea $X_1, X_2, \dots, X_n$ una sucesión de variables aleatorias **independientes e idénticamente distribuidas (i.i.d.)** con $\mathbb{E}[X_i] = \mu$.

Entonces, conforme $n \to \infty$:

$$
\bar{X}_n = \frac{1}{n}\sum_{i=1}^n X_i \xrightarrow{p} \mu
$$

Esto significa que la **media muestral converge en probabilidad** a la media poblacional.


#### Interpretación

- Es el fundamento de la **estimación estadística**.
- Justifica que la **media de muchas observaciones** aproxima el valor verdadero esperado.


#### Ejemplo (Python)

Simulación con una distribución uniforme:

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación
np.random.seed(42)
n = 1000
X = np.random.uniform(0, 1, n)
media_muestral = np.cumsum(X) / np.arange(1, n + 1)

# Visualización
plt.figure(figsize=(7, 4))
plt.plot(media_muestral, label='Media muestral acumulada')
plt.axhline(0.5, color='red', linestyle='--', label='Media teórica μ = 0.5')
plt.title("Ley de los Grandes Números (Uniforme(0,1))")
plt.xlabel("n")
plt.ylabel("Media acumulada")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

## Teorema Central del Límite (TCL)

#### Enunciado

Sea $X_1, X_2, \dots, X_n$ una secuencia de variables i.i.d. con media $\mu$ y varianza $\sigma^2$.  
Entonces, la variable estandarizada:

$$
Z_n = \frac{\bar{X}_n - \mu}{\sigma / \sqrt{n}}
$$

converge en distribución a una **Normal Estándar**:

$$
Z_n \xrightarrow{d} N(0,1)
$$


#### Interpretación

El TCL explica por qué **la suma o promedio de variables aleatorias** tiende a una distribución normal, independientemente de su distribución original.


#### Ejemplo (Python)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(1)
n = 10000
muestras = 1000

# Generamos n variables uniformes y calculamos medias de tamaño m
medias = []
for m in [2, 5, 10, 30]:
    for _ in range(muestras):
        X = np.random.uniform(0, 1, m)
        medias.append(((np.mean(X) - 0.5) / (np.sqrt(1/12) / np.sqrt(m))))

# Graficamos la aproximación a la normal estándar
plt.figure(figsize=(7, 4))
plt.hist(medias, bins=50, density=True, color='lightblue', edgecolor='black', alpha=0.6)
x = np.linspace(-4, 4, 200)
plt.plot(x, norm.pdf(x, 0, 1), 'r', lw=2, label='Normal(0,1)')
plt.title("Teorema Central del Límite — Uniforme(0,1)")
plt.xlabel("Z estandarizado")
plt.ylabel("Densidad")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
