## Espacio de probabilidad  



### Definicion

La teoría moderna de la probabilidad, basada en los **axiomas de Kolmogórov (1933)**, formaliza la noción de azar mediante la definición de un **espacio de probabilidad**.

Un **espacio de probabilidad** se define como una **triple**:

$$
(\Omega, \mathcal{F}, P)
$$

donde:

- **$\Omega$** es el **espacio muestral**: el conjunto de todos los posibles resultados de un experimento aleatorio.  
  Ejemplo: si lanzamos un dado,  
  $$
  \Omega = \{1, 2, 3, 4, 5, 6\}.
  $$

- **$\mathcal{F}$** es una **σ–álgebra** (sigma–álgebra) de subconjuntos de $\Omega$.  
  Es decir, un conjunto de eventos tal que:
  1. $\Omega \in \mathcal{F}$,  
  2. Si $A \in \mathcal{F}$, entonces $A^c \in \mathcal{F}$,  
  3. Si $A_1, A_2, A_3, \dots \in \mathcal{F}$, entonces $\bigcup_{i=1}^{\infty} A_i \in \mathcal{F}$.

  Cada elemento de $\mathcal{F}$ se denomina **evento**.

- **$P$** es una **función de probabilidad**,  
  $$
  P: \mathcal{F} \to [0,1],
  $$
  que asigna a cada evento $A \in \mathcal{F}$ un número real $P(A)$, llamado **probabilidad del evento $A$**.

### Axiomas de Kolmogórov

La función de probabilidad $P$ debe satisfacer los siguientes **tres axiomas fundamentales**:

1. **No negatividad**
   $$
   P(A) \geq 0, \quad \forall A \in \mathcal{F}.
   $$

2. **Normalización**
   $$
   P(\Omega) = 1.
   $$

3. **Aditividad numerable**
   Si $A_1, A_2, A_3, \dots$ son eventos mutuamente excluyentes  
   (es decir, $A_i \cap A_j = \emptyset$ para $i \neq j$), entonces:
   $$
   P\left(\bigcup_{i=1}^{\infty} A_i\right) = \sum_{i=1}^{\infty} P(A_i).
   $$

#### Consecuencias inmediatas de los axiomas

A partir de los axiomas se deducen propiedades básicas:

1. $P(\emptyset) = 0$  
2. Para cualquier evento $A$,  
   $$
   P(A^c) = 1 - P(A)
   $$
3. Si $A \subseteq B$, entonces  
   $$
   P(A) \leq P(B)
   $$
4. Para dos eventos cualesquiera $A$ y $B$:
   $$
   P(A \cup B) = P(A) + P(B) - P(A \cap B)
   $$

Los axiomas de Kolmogórov permiten que la probabilidad se interprete como una medida sobre el espacio de eventos, dotando a la teoría de una base matemáticamente rigurosa.  
De esta forma, la probabilidad deja de ser una noción intuitiva y pasa a integrarse dentro del marco de la teoría de la medida, garantizando coherencia en todos los resultados posteriores de la estadística matemática.

## Enfoque Clásico de la Probabilidad

### Definicion

El **enfoque clásico** o **laplaciano** de la probabilidad parte de la idea de que todos los resultados de un experimento aleatorio son **igualmente probables**.  
Este enfoque es apropiado cuando el número de resultados posibles es **finito y equiprobable** (como en dados, monedas o cartas).


### Espacio muestral $(\Omega)$

El **espacio muestral** es el conjunto de todos los resultados posibles de un experimento aleatorio.

Se define como: 
$$
\Omega = \{\omega_1, \omega_2, \dots, \omega_n\}
$$

Ejemplo:  
Si lanzamos un dado,
$$
\Omega = \{1,2,3,4,5,6\}.
$$


### Eventos


Un **evento** es cualquier subconjunto del espacio muestral $( \Omega )$.

Si $( A \subseteq \Omega )$, decimos que $A$ es un evento.  
- Si el resultado pertenece a $A$, el evento ocurre.  
- Si no pertenece, el evento no ocurre.

**Ejemplo:**  
Sea $A = \{\text{salir un número par}\} = \{2,4,6\}$.


### Probabilidad (en el enfoque clásico)

Si el experimento tiene $n$ resultados igualmente probables y un evento $A$ ocurre en $m$ de ellos, entonces:

$$
P(A) = \frac{m}{n}
$$

donde:
- $m$ = número de resultados favorables a $A$  
- $n$ = número total de resultados posibles

**Ejemplo:**  
En un dado, la probabilidad de obtener un número par es:

$$
P(A) = \frac{3}{6} = \frac{1}{2}
$$


### Probabilidad condicional

La **probabilidad condicional** mide la probabilidad de que ocurra un evento $A$, dado que ya ocurrió otro evento $B$:

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)}, \quad \text{si } P(B) > 0
$$

**Interpretación:**  
Se restringe el espacio muestral original a la condición de que $B$ haya ocurrido.

**Ejemplo:**

Se lanzan **dos dados**. Sea:
- $A$ = “la suma de los dados es 8”,  
- $B$ = “la suma de los dados es par”.

El espacio muestral tiene $6 \times 6 = 36$ resultados posibles.

Los resultados donde la **suma es par** $(B)$ son 18:
$$
B = \{(1,1),(1,3),(1,5),(2,2),(2,4),(2,6), \dots, (6,6)\}
$$
por lo tanto:
$$
P(B) = \frac{18}{36} = \frac{1}{2}.
$$

Los resultados donde la **suma es 8** $(A)$ son 5:
$$
A = \{(2,6),(3,5),(4,4),(5,3),(6,2)\}
$$
y todos estos también son pares, así que $A \subset B$, lo que implica:
$$
P(A \cap B) = P(A) = \frac{5}{36}.
$$

Entonces:
$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)} = \frac{\frac{5}{36}}{\frac{1}{2}} = \frac{5}{18}.
$$

**Respuesta:**  
La probabilidad de que la suma sea 8 sabiendo que la suma fue par es:

$$
P(A \mid B) = \frac{5}{18} \approx 0.2778.
$$



### Independencia de eventos

Dos eventos $A$ y $B$ son **independientes** si la ocurrencia de uno **no afecta** la probabilidad del otro.  
Formalmente:

$$
P(A \cap B) = P(A) \, P(B)
$$

Equivalente a decir que:

$$
P(A \mid B) = P(A) \quad \text{y} \quad P(B \mid A) = P(B)
$$

**Ejemplo:**  
Al lanzar **dos monedas**, definimos los eventos:
- $A = \{\text{la primera moneda es cara}\}$
- $B = \{\text{la segunda moneda es cara}\}$

El espacio muestral es:
$$
\Omega = \{(C,C), (C,S), (S,C), (S,S)\}
$$

Entonces:
$$
P(A) = \frac{2}{4} = \frac{1}{2}, \quad P(B) = \frac{2}{4} = \frac{1}{2},
$$
y
$$
P(A \cap B) = \frac{1}{4}.
$$

Como:
$$
P(A \cap B) = P(A)P(B) = \frac{1}{2} \cdot \frac{1}{2} = \frac{1}{4},
$$
los eventos $A$ y $B$ son **independientes**.

El enfoque clásico, aunque limitado a experimentos equiprobables, sienta las bases intuitivas para el desarrollo axiomático moderno de la probabilidad.

```python
## Ejercicios para complementar de forma práctica la estructura de un espacio de probabilidad, comprobar los axiomas de Kolmogórov, y aplicar la probabilidad condicional y la independencia

import itertools
import numpy as np

# 1) Espacio muestral (Ω) — Lanzamiento de dos dados justos
# Ω: conjunto de todos los pares posibles (i, j), donde i y j ∈ {1,...,6}
Omega = list(itertools.product(range(1, 7), repeat=2))
n_total = len(Omega)
print("Espacio muestral Ω =", Omega)
print("Número total de resultados posibles |Ω| =", n_total)

# 2) Definición de eventos (subconjuntos de Ω)
# A: la suma de los dados es 7
# B: el primer dado muestra un número par
# C: la suma de los dados es par

A = [(i, j) for (i, j) in Omega if i + j == 7]
B = [(i, j) for (i, j) in Omega if i % 2 == 0]
C = [(i, j) for (i, j) in Omega if (i + j) % 2 == 0]

# 3) Definición de la función de probabilidad (P)
# En el enfoque clásico, todos los resultados son equiprobables:
P = lambda evento: len(evento) / n_total

print("\nProbabilidades básicas:")
print("P(A) = P(suma=7) =", P(A))
print("P(B) = P(primer dado par) =", P(B))
print("P(C) = P(suma par) =", P(C))

# 4) Verificación de los axiomas de Kolmogórov
# 1. No negatividad
assert all(P(E) >= 0 for E in [A, B, C]), "Violación del axioma de no negatividad."

# 2. Normalización
assert np.isclose(P(Omega), 1.0), "Violación del axioma de normalización."

# 3. Aditividad para eventos disjuntos
# Ejemplo: evento D1 = {(1,1)} y D2 = {(1,2)}
D1 = [(1, 1)]
D2 = [(1, 2)]
union = D1 + D2
assert np.isclose(P(D1) + P(D2), P(union)), "Violación del axioma de aditividad."

print("\nVerificación de axiomas de Kolmogórov: todos los axiomas se cumplen.")

# 5) Probabilidad condicional: P(A | C)
# A: suma = 7, C: suma par
interseccion = [w for w in A if w in C]
P_A_dado_C = P(interseccion) / P(C)
print("\nProbabilidad condicional:")
print("P(A|C) = P(A∩C)/P(C) =", P_A_dado_C)

# 6) Independencia de eventos: A y B
# Verificamos si P(A∩B) = P(A)*P(B)
inter_AB = [w for w in A if w in B]
P_AcapB = P(inter_AB)
independencia = np.isclose(P_AcapB, P(A) * P(B))

print("\nIndependencia:")
print("P(A∩B) =", P_AcapB)
print("P(A)*P(B) =", P(A) * P(B))
print("¿A y B son independientes? →", independencia)
```

## Enfoque Geométrico de la Probabilidad


### Definicion

El **enfoque geométrico** de la probabilidad surge como una extensión natural del enfoque clásico a situaciones donde los resultados posibles **no pueden contarse**, sino que forman un **continuo** dentro de una región del espacio.

En este caso, los eventos y el espacio muestral se describen por medio de **longitudes**, **áreas** o **volúmenes**, y la probabilidad se define como una **razón de medidas**.

Sea $(\Omega, \mathcal{F}, P)$ un espacio de probabilidad, donde el espacio muestral $\Omega$ corresponde a una región geométrica medible.

Si todos los puntos de $\Omega$ son **igualmente probables**, y $A \subseteq \Omega$ es un evento representado por una subregión del mismo tipo, la **probabilidad geométrica** se define como:

$$
P(A) = \frac{\text{medida de } A}{\text{medida de } \Omega}
$$

donde “medida” puede ser:
- Longitud (si el experimento es unidimensional),
- Área (si el experimento ocurre en el plano),
- Volumen (si el experimento ocurre en el espacio tridimensional).

### Interpretación

La probabilidad de que un punto elegido **al azar** en $\Omega$ pertenezca al conjunto $A$ es proporcional al tamaño relativo de $A$ respecto a $\Omega$.

Si todos los puntos de $\Omega$ son igualmente probables, la probabilidad depende únicamente del **razonamiento geométrico** (medidas, proporciones y áreas).

Ejemplo 1 — Punto en un segmento

Supongamos que un punto se elige **al azar** sobre un segmento de longitud $L=10$.

Sea el evento $A$ = “el punto cae entre las posiciones 3 y 7”.

Entonces:
$$
P(A) = \frac{\text{longitud de } A}{\text{longitud total}} = \frac{7-3}{10} = \frac{4}{10} = 0.4
$$

**Interpretación:**  
La probabilidad de que el punto caiga entre 3 y 7 es 0.4, porque el subsegmento $[3,7]$ ocupa el 40% del segmento total.

Ejemplo 2 — Punto en un cuadrado

Sea $\Omega$ un cuadrado de lado 4, dentro del cual se elige **al azar** un punto $(x, y)$.

Sea $A$ el evento “el punto está dentro del círculo de radio 2 centrado en el mismo punto que el cuadrado”.

El área del cuadrado es:
$$
A_\Omega = 4^2 = 16
$$

El área del círculo es:
$$
A_A = \pi r^2 = \pi (2)^2 = 4\pi
$$

Entonces:
$$
P(A) = \frac{A_A}{A_\Omega} = \frac{4\pi}{16} = \frac{\pi}{4} \approx 0.7854
$$

**Interpretación:**  
Si se elige un punto aleatorio dentro del cuadrado, la probabilidad de que esté dentro del círculo es aproximadamente 0.785.

Ejemplo 3 — Aguja de Buffon (probabilidad geométrica continua)

Un ejemplo clásico del enfoque geométrico es el **problema de la aguja de Buffon**, utilizado incluso para estimar el número $\pi$.

Una aguja de longitud $l$ se deja caer sobre un plano con líneas paralelas separadas por una distancia $d > l$.

La probabilidad de que la aguja toque alguna línea es:

$$
P = \frac{2l}{\pi d}
$$

Este problema une la **probabilidad geométrica** con la **medida angular** y la **integración continua**.


### Relación con el enfoque clásico y axiomático

- Cuando los resultados son **discretos y equiprobables**, se usa el **enfoque clásico**:
  $$
  P(A) = \frac{m}{n}
  $$

- Cuando los resultados forman un **continuo uniforme** (como longitudes o áreas), se usa el **enfoque geométrico**:
  $$
  P(A) = \frac{\text{medida de }A}{\text{medida de }\Omega}
  $$

- En ambos casos, la definición respeta los **axiomas de Kolmogórov**:
  1. $P(A) \ge 0$
  2. $P(\Omega) = 1$
  3. Si $A_i$ son disjuntos, $P(\bigcup A_i) = \sum P(A_i)$

El **enfoque geométrico** extiende la idea de “equiprobabilidad” al caso continuo, donde los resultados ya no se cuentan sino que se **miden**.  
Es la base para definir **distribuciones continuas**, como la **Uniforme continua**, la **Exponencial** y la **Normal**.

### El problema del encuentro

Romeo y Julieta pretenden encontrarse en un lugar específico entre $[0,60]$

- Si Romeo llega primero, entonces piensa esperar 10 minutos
- Si Julieta llega primero, entonces piensa esperar 10 minutos

Se elige un número al azar entre 0 y 60, y supongamos que el par de números se elige sobre el cuadrado de modo que la probabilidad de un evento sea el área favorable entre el área total

Sea el evento
- $A$: Romeo y Julieta se encuentran a una distancia de 10 minutos

```python
import numpy as np
import matplotlib.pyplot as plt

# Vamos a generar 100,000 simulaciones 
muestra = 100000

# Generamos los tiempos aleatorios entre [0,60]
# x : tiempo de llegada de Romeo
# y : tiempo de llegada de Julieta

x = np.random.uniform(0, 60, muestra) # genera 100,000 números aleatorios entre 0 y 60
y = np.random.uniform(0, 60, muestra) # genera 100,000 números aleatorios entre 0 y 60

# Vamos a añadir mi condición de encuentro
encuentro = np.where(abs(x-y) <=10, 'r', 'b') 
# Si la diferencia entre los tiempos es menor o igual a 10, entonces se encuentran (rojo), 
# si no, no se encuentran (azul)

# Vamos a calcular la probabilidad de que se encuentren
k = 0
for elemento in encuentro:
    if elemento == 'r':
        k += 1

print("La probabilidad del evento A es: {}.".format(k/muestra))

# Para graficar los puntos de encuentro, hacemos una gráfica de dispersión
plt.figure(figsize=(8,8))
plt.scatter(x, y, c=encuentro, s=1)
plt.show()
```

## Enfoque Frecuentista de la Probabilidad

### Definicion

El **enfoque frecuentista** interpreta la probabilidad como el **límite de la frecuencia relativa** con que ocurre un evento cuando se repite el experimento un gran número de veces bajo condiciones idénticas.

Este enfoque se fundamenta en la **observación empírica** y es uno de los pilares de la **estadística inferencial**.

Sea $A$ un evento asociado a un experimento aleatorio, y sea $n(A)$ el número de veces que ocurre el evento $A$ después de realizar el experimento $n$ veces.

La **frecuencia relativa** de $A$ está dada por:

$$
f_n(A) = \frac{n(A)}{n}
$$

El **enfoque frecuentista** define la probabilidad de $A$ como el **límite de esa frecuencia relativa** cuando el número de repeticiones tiende a infinito:

$$
P(A) = \lim_{n \to \infty} f_n(A) = \lim_{n \to \infty} \frac{n(A)}{n}
$$


### Interpretación

En este enfoque, la probabilidad **no se asume a priori**, sino que se **estima** a partir de la observación de los resultados en muchos experimentos repetidos.

Por tanto:
- La probabilidad refleja **estabilidad a largo plazo**.
- No se habla de “probabilidad” de un suceso único, sino del **comportamiento promedio** a medida que el número de repeticiones aumenta.


### Ejemplo — Volado (moneda equilibrada)

Supongamos que lanzamos una moneda muchas veces.  
Definimos el evento:
- $A =$ “obtener cara”.

En cada lanzamiento, el resultado puede ser **cara (C)** o **cruz (S)**.  
Si la moneda es justa, esperamos que en promedio $P(A) = 0.5$, pero esta probabilidad se **observa** al repetir el experimento.

La frecuencia relativa después de $n$ lanzamientos es:

$$
f_n(A) = \frac{n(A)}{n}
$$

donde:
- $n(A)$ es el número de veces que salió **cara**,
- $n$ es el número total de lanzamientos.

### Ejemplo Numerico

Supongamos los siguientes resultados:

| Número de lanzamientos | Veces que salió cara | Frecuencia relativa |
|------------------------:|---------------------:|--------------------:|
| 10 | 6 | 0.60 |
| 100 | 48 | 0.48 |
| 1,000 | 511 | 0.511 |
| 10,000 | 4,974 | 0.4974 |

Podemos observar que conforme aumenta el número de lanzamientos,  
la frecuencia relativa **se aproxima a 0.5**, valor teórico esperado.

Esto ilustra que:

$$
\lim_{n \to \infty} \frac{n(A)}{n} = 0.5
$$

El enfoque frecuentista:
- Define la probabilidad como **una medida empírica** de la estabilidad de las frecuencias.
- Depende del **repetir el experimento** un número suficientemente grande de veces.
- Es la base para las **estimaciones estadísticas** y la **inferencia basada en muestreo**.

| Enfoque | Característica principal | Ejemplo |
|----------|-------------------------|----------|
| Clásico | Equiprobabilidad de resultados | Dado o ruleta |
| Geométrico | Razón entre medidas (longitud, área, volumen) | Punto al azar en un cuadrado |
| **Frecuentista** | Límite de frecuencias observadas | Volado repetido |
| Axiomático | Fundamento formal con propiedades abstractas | Espacios de probabilidad |


```python
##Simulación del enfoque frecuentista de la probabilidad

##Experimento: lanzar una moneda equilibrada (cara = 1, cruz = 0)
##y observar cómo la frecuencia relativa de "cara" se estabiliza
##alrededor de 0.5 cuando n → ∞.

import numpy as np
import matplotlib.pyplot as plt


# 1. Definición del experimento
# Lanzamientos totales
N = 10_000

# Semilla para reproducibilidad
np.random.seed(42)

# Simulación de lanzamientos (0 = cruz, 1 = cara)
resultados = np.random.randint(0, 2, N)

# 2. Cálculo de la frecuencia relativa acumulada
# np.cumsum: suma acumulada de caras (1s)
caras_acumuladas = np.cumsum(resultados)

# Frecuencia relativa después de cada lanzamiento
frecuencia_relativa = caras_acumuladas / np.arange(1, N + 1)

# 3. Visualización
plt.figure(figsize=(8, 4))
plt.plot(frecuencia_relativa, color='darkorange', lw=1.8, label='Frecuencia relativa de "cara"')
plt.axhline(0.5, color='black', linestyle='--', label='Valor teórico: 0.5')
plt.title('Enfoque Frecuentista de la Probabilidad — Volado', fontsize=12)
plt.xlabel('Número de lanzamientos')
plt.ylabel('Frecuencia relativa de "cara"')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# 4. Resultado final
print(f"Frecuencia relativa final (n={N}): {frecuencia_relativa[-1]:.4f}")
```
